```
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
```

```
iris = load_iris()
attr = iris.data
label = iris.target

x_train, x_test, y_train, y_test = train_test_split(
    attr, label, test_size=0.2)
```

Decision Tree without PCA

```
model = DecisionTreeClassifier()
pertama = model.fit(x_train, y_train)
pertama.score(x_test, y_test)
```

Apply PCA pada dataset

```
pca = PCA(n_components=4)
pca_attr = pca.fit_transform(x_train)
pca.explained_variance_ratio_
```

```
pca = PCA(n_components=2)
x_train_pca = pca.fit_transform(x_train)
x_test_pca = pca.fit_transform(x_test)

kedua = model.fit(x_train_pca, y_train)
kedua.score(x_test_pca, y_test)
```
